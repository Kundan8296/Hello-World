```python
#Q-Q plot
import numpy as np
import pylab
import scipy.stats as stats
```

```python
#step-1 create a random varaible which is normally distributed-std_normal
std_normal=np.random.normal(loc=0,scale=1,size=1000)

# 0 to 100th percentiles of std-normal
for i in range(0,101):
    print(i, np.percentile(std_normal,i))
```

```python
#craete another random varale which is to be tested -X
X=np.random.normal(loc=20,scale=5,size=1000)

stats.probplot(X,dist='norm',plot=pylab)

pylab.show()
```

```python
#another one to be tested-Z
Z=np.random.uniform(low=-1,high=1,size=1000)
stats.probplot(Z,dist="norm",plot=pylab)
pylab.show()
```
